Write a program that:

вЂўвЃ   вЃ Demonstrates break, continue, and pass

```python
for i in range(1,11):
    if i==3:
        continue
    if i==5:
        pass
    if i==7:
        break
    print(i)
```

#### 7.вЃ  вЃ Recursion


Write a recursive function that:

вЂўвЃ   вЃ Finds the factorial of a number


```python
def fact(n):
    if n==0:
        return 1
    else:
        return n*fact(n-1)
print(fact(10))
```
