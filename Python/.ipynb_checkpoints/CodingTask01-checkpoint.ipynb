#### 2. Even or Odd


```python
num = int(input("Enter the number : "))
if num%2==0:
    print("It's an even number.")
else:
    print("It's an odd number.")
```

#### 3. List Operations

```python
nums=[10,20,30,40,50]
print("Sum = " ,sum(nums))
print("Max = " ,max(nums))
print("Min = " ,min(nums))  
```
